```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Mobiles_Dataset.csv', encoding="ISO-8859-1") #utf8 olarak  okumaya çalışır  bizim dosyamız ise latin1 formatında
```

```python
print("Veri setinin boyutu: ", df.shape) # df'in boyutu
```

```python
print("Sütun isimleri:", df.columns.to_list()) 
```

```python
print("Seçilen sürunlarınn standart sapması:")
print(df.loc[:,['Launched Year']].std()) # sadece yıl ı alabildik çünkü diğer sayısal verilerin içerisinde ya harf ya da yazı olduğu için string olarak algılanıyor.
#loc sütun adı ile çalışır loc=location
```

```python
print("Yıl Sütununun ortalaması: ")
print(df.iloc[:,14:15].mean()) #14.sütundan 15.sütuna kadar anlamına gelir
#iloc= index location: sütunun sırasını kullanarak seçim yapılır
```

```python
filtreli_veriler=df[(df['Launched Year'] > 2015) & (df['Launched Year'] < 2020)]
print("2015-2020 arasında olan telefonlar")
print(filtreli_veriler)
# 'Launched Year' sütunundaki 2015-2020 arasında kalan veri sayısı
```
